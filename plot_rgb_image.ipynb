```python
def _add_plot_text( fig, label, blorder, distance):
    # set nan color 
    fig.set_nan_color('0.95')
    
    # beam 
    fig.add_beam()
    fig.beam.set_color('#E31A1C')
    
    # Scale bar
    ang_sep = (1000*u.au/distance)*u.rad
    fig.add_scalebar(ang_sep)
    fig.scalebar.set_label('1000 au')
    fig.scalebar.set_font(size=22)
    fig.scalebar.set_linewidth(4) 
    
    # label
    fig.add_label(0.90,0.95, label, relative=True, color='black', size=40)
    
    # Add region and tick labels parameters
    fig.tick_labels.set_xformat('d.dd')
    fig.tick_labels.set_yformat('d.dd')
    fig.tick_labels.set_font(size=26)
    fig.ticks.set_color('black')
    fig.ticks.set_minor_frequency(4)
    fig.ticks.set_xspacing(0.04)  # degrees
        
    fig.axis_labels.set_font(size=26)
    
```

```python
rms = 0.017
c_levs = [3.*rms, 6.*rms, 12.*rms, 24.*rms, 48.*rms, 96.*rms]

w11_full = '/Users/kristina/NH3/ammonia_project/data/moment0_11_Jy.fits'
fig = plt.figure(figsize=(30,20), dpi=300)

####
#### Tk
####
color_table='autumn_r'
tk_plot = aplpy.FITSFigure(hdu, dimensions=[0,1], slices=[0], figure=fig, subplot=(1,3,1))
tk_plot.show_colorscale(cmap=color_table, vmin=2.8, vmax=32)
tk_plot.show_contour(w11_full, colors='black', linewidths=1.5, levels=c_levs)
_add_plot_text(tk_plot, label='$T_{\mathrm{K}}$', blorder=1, distance=388*u.pc)

tk_plot.add_colorbar()
tk_plot.colorbar.set_location('top')
tk_plot.colorbar.set_width(0.2)
tk_plot.colorbar.set_pad(0.)
#tk_plot.colorbar.set_axis_label_pad(1.)
tk_plot.colorbar.set_axis_label_text('(K)')
tk_plot.colorbar.set_axis_label_font(family='sans_serif',size=26)
tk_plot.colorbar.set_font(family='sans_serif',size=26)

####
#### Tex
####
color_table='autumn_r'
tex_plot = aplpy.FITSFigure(hdu, dimensions=[0,1], slices=[1], figure=fig, subplot=(1,3,2))
tex_plot.show_colorscale(cmap=color_table, vmin=2.8, vmax=17)
tex_plot.show_contour(w11_full, colors='black', linewidths=1.5, levels=c_levs)
_add_plot_text(tex_plot, label='$T_{\mathrm{ex}}$', blorder=1, distance=388*u.pc)

tex_plot.add_colorbar()
tex_plot.colorbar.set_location('top')
tex_plot.colorbar.set_width(0.2)
tex_plot.colorbar.set_pad(0.)
#tex_plot.colorbar.set_axis_label_pad(1.)
tex_plot.colorbar.set_axis_label_text('(K)')
tex_plot.colorbar.set_axis_label_font(family='sans_serif',size=26)
tex_plot.colorbar.set_font(family='sans_serif',size=26)
tex_plot.hide_ytick_labels()
tex_plot.hide_yaxis_label()

####
#### N(NH3)
####
color_table='winter'
nnh3_plot=aplpy.FITSFigure(hdu, dimensions=[0,1], slices=[2], figure=fig, subplot=(1,3,3))
nnh3_plot.show_colorscale( cmap=color_table, vmin=14.25, vmax=15.25)
nnh3_plot.show_contour(w11_full, colors='black', linewidths=1.5, levels=c_levs)
_add_plot_text( nnh3_plot, label='', blorder=1, distance=388*u.pc)
nnh3_plot.add_label(0.75,0.95, '$\log(N$(NH$_3$)/cm$^{-2}$)', relative=True, color='black', size=30)

nnh3_plot.add_colorbar()
nnh3_plot.colorbar.set_location('top')
nnh3_plot.colorbar.set_width(0.2)
nnh3_plot.colorbar.set_pad(0.)
nnh3_plot.colorbar.set_axis_label_font(family='sans_serif',size=26)
nnh3_plot.colorbar.set_font(family='sans_serif',size=26)
nnh3_plot.hide_ytick_labels()
nnh3_plot.hide_yaxis_label()


fig.subplots_adjust(wspace=0.)

fig.savefig('/Users/kristina/NH3/ammonia_project/plots/parameter_maps/snr1/fit_temp.pdf', bbox_inches='tight')
```
